# Testing the model

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
model = load_model('ECG.h5') #loading our trained model
```

```python
# test model 0
img=image.load_img('/content/data/test/Left Bundle Branch Block/fig_5935.png',target_size=(64,64))#loading image
```

```python
import numpy as np
x= image.img_to_array(img)
x=np.expand_dims(x,axis=0)
```

```python
x.shape
```

```python
preds=model.predict(x)
pred=np.argmax(preds,axis=1)
preds
```

```python
pred[0]
```

```python
a=model.predict(x)
```

```python
index=['Left Bundle Branch Block','Normal','Premature Atrial Contraction',
       'Premature Ventricular Contractions', 'Right Bundle Branch Block','Ventricular Fibrillation']
result=str(index[pred[0]])
result
```
